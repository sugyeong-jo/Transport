```python
busstop_hs = busstop[busstop.iloc[:,6]=="경기도 화성시"]
busstop_hs
```

```python
hs_idx = list(set(busstop_hs.iloc[:,7].dropna(axis=0)))
print(len(hs_idx))
print(hs_idx)
```

```python
hs_info = dict()
for k in hs_idx:
    tmp = busstop[busstop.iloc[:,7]==k]
    tmp_busstop = list(set(tmp.iloc[:,0]))
    ## print # of busstop based on (busstop id, busstop name, busstop number)
    print(k,": ", len(list(set(tmp.iloc[:,0]))), len(list(set(tmp.iloc[:,2]))), len(list(set(tmp.iloc[:,4])))) 
    cnt = 0
    for lane in tmp_busstop:
        if cnt == 0:
            sel_dat = bus_route[bus_route.iloc[:,5]==lane]
            cnt += 1
        else:
            sel_dat = pd.concat([sel_dat, bus_route[bus_route.iloc[:,5]==lane]])
            cnt += 1
    hs_info[k] = sel_dat
print(len(hs_idx), len(hs_info))
```

```python
for k in hs_info.keys():
    print(k,": ", len(hs_info[k]))
```
